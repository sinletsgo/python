```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT * FROM customerdata', engine) 

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']

# 데이터 VIEW
customerData.head()
```

### 데이터프레임명[‘컬럼명‘] = np.where( 조건식, yes값, no값)

```python

# numpy 라이브러리 
import numpy as np

#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
customerData["PRODUCTAGE_NEW"] = np.\
where(customerData["PRODUCTAGE"] < 1, 1, customerData["PRODUCTAGE"])


customerData.head()
```

### quiz
productage가 1보다작으면 1 2보다작으면2 3보다작으면3
이외는 5라고 새로운 컬럼 product_age_new를 생성하세요


```python
#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
# 1,2,3
# step = [1,2,3]

step = list(range(1,4))
maxValue =5



customerData["PRODUCTAGE_NEW"] = np.where(customerData["PRODUCTAGE"] < step[0], step[0],\
                                          np.where(customerData["PRODUCTAGE"] < step[1], step[1],\
                                          np.where(customerData["PRODUCTAGE"] < step[2], step[2],\
                                          maxValue)))
customerData.head()



```

## 이상 데이터 정제하기 #2


```python

#컬럼 추가 생성 및 조건에 따른 컬럼명 변경
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5

customerData["productage_new2"] =customerData.apply(refining, axis=1)

customerData.head()
```

```python

spark에서 slide 했던 것과 비슷한것이 여기도 있다.
center 평균구할때 사용, window 5개씩 롤링하며 평균.

자신 포함해서 5개. 즉 양 날개 2개씩 포함해서 TEST 컬럼에 평균치를 낸다.

응용하면
stock price 보는데, 단기, 장기 이동펻균의 차이를 구하고,
10주 평균 1천, 5주 평균 2천이였다?
최근에 잘 팔린거지!

단기 - 장기 = 1보다 크면 사야겠다

식으로 구할 때 rolling 사용!
```

### 맨 위 2개는 5개 데이터 확보가 안되서 안채워진다. 이것도 채우려면 아래와 같이.

```python
import math

suborder = math.floor(5/2)
suborder
```

```python
# 즉 맨위 2개 데이터에 대한 PRODUCTAGE 평균만 구한걸 list에 넣어서 넣어준거다.
```

```python
list = []
for i in range(0, suborder):
    list.append(customerData["PRODUCTAGE"][0:i+ suborder].mean())

list
```

```python
for i in range(0, suborder):
    customerData["TEST"][i] = list[i]
```
