# 이동평균 rolling

```python

spark에서 slide 했던 것과 비슷한것이 여기도 있다.
center 평균구할때 사용, window 5개씩 롤링하며 평균.

window= 5개 하면, 자신 포함해서 5개. 즉 양 날개 2개씩 포함해서 TEST 컬럼에 평균치를 낸다.

응용하면
stock price 보는데, 단기, 장기 이동펻균의 차이를 구하고,
10주 평균 1천, 5주 평균 2천이였다?
최근에 잘 팔린거지!

단기 - 장기 = 1보다 크면 사야겠다
식으로 구할 때 rolling 사용!
```

```python
customerData1["TEST"] = customerData1["PRODUCTAGE"].rolling(window = 5, center =False).mean()
```

```python
customerData1.head(10)
```

```python
customerData1["PRODUCTAGE"][1:6].mean()
```

```python
center =True 로 하면 자기 자신이 중심이 되서 양 날개 5개의 합계 평균을 구한다
```

```python
customerData1["TEST"] = customerData1["PRODUCTAGE"].rolling(window = 5, center =True).mean()
```
