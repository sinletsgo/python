```python
# 5가 센터를 중심으로 2씩 떨어진다!. 
# 중간 값 계산.
import math
# 이동평균 구간
order = 5
# 앞뒤로 비어있게될 구간 크기
suborder = math.floor(order/2)
suborder
```

### 아래쪽 채우기

```python
sortedData["VOLUME"][ (maxLength- 1)- (suborder+i):maxLength].mean() 
```

```python
postList = []
maxLength = len(sortedData)


for i in range(0, suborder):
    postList.append( sortedData["VOLUME"][ (maxLength- 1)- (suborder+i):maxLength].mean()  )
    sortedData.loc[maxLength-(i+1), "MA"] = postList[i] 

postList
```

```python
sortedData.tail()
```

### 위쪽 채우기

```python
preList = []

for i in range(0, suborder):
    preList.append( sortedData["VOLUME"][ 0:suborder+i +1].mean()  )
    sortedData.loc[i, "MA"] = preList[i] 

preList 
```

```python
sortedData.head()
```

### ratio 구하기

```python
sortedData["RATIO"] = sortedData["VOLUME"]/sortedData["MA"]

sortedData.head()
```

# BUT

```python
하지만 이동평균이 잘 구해진것 같지만 롤링이 제대로 안된거다.

이동평균을 그룹핑안하고 돌리니까, ST0001 -> ST0002로 넘어갈때, 갑자기 값이 달라진다.
제대로 된 PRODUCT별 이동평균 정보가 아닌것!

GROUR BY 해서 롤링하는 방법을 알아야 한다.
```

```python
def sub_function(data):
    data["MA"] = data["VOLUME"].rolling(window = 5, center =True).mean()
    return data
```

```python
# apply 적용하면 groupby 된 data 가 자동으로 함수 전달인자로 들어가나 보네..
groupResult = sortedData.groupby(["REGIONID", "PRODUCTGROUP"]).apply(sub_function)
groupResult.head()
```
