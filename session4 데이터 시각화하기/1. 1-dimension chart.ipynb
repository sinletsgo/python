### 2차원 차트 그리기

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT * FROM customerdata', engine) 

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']
customerData.head(2)
```

```python
customerData["CAT_CUST"] = customerData.CUSTTYPE.astype("category").cat.codes
```

```python
customerData["CUSTTYPE_NEW"] = np.where(customerData["CUSTTYPE"] == "Big-Screen-lover", 1,\
                                         np.where(customerData["CUSTTYPE"] == "Sleeping-dog", 2,\
                                         np.where(customerData["CUSTTYPE"] == "Early-bird", 3,\
                                         4)))
customerData.head()
```

### 분포도 차트 구현

```python
#pip install seasborn
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
sns.distplot(tips["total_bill"])
```

### jointplot구현

```python
sns.jointplot(x="total_bill", y = "tip", data = tips)
```

### pair plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = "sex")
```

### barplot 차트

```python
sns.barplot(x="sex", y='total_bill', data = tips)
```

### count plot

```python
sns.countplot(x= 'sex', data = tips)
```

### boxplot 구현`

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```
