## 1차원 차트 그리기

x,y축을 바탕으로 차트 시각화


```python
#필요 라이브러리 정의
import numpy as np # 숫자배열연산 라이브러리 호출

import matplotlib.pyplot as plt # 차트 용 라이브러리 호출


# 커맨드뷰에서 차트 시연
%matplotlib inline
# 팝업 창 활용하여 차트 시연
# %matplotlib tk
size = 50
#사이즈만큰 랜덤한 데이터 생성 (정규분포(standard_normal): 평균0, 표준편차1)
y=np.random.standard_normal(size)
x=range(size)
plt.plot(x,y)
```

## 1차원 차트 꾸미기

차트의 색 및 내용 채우기


```python
==========  ========
character   color
==========  ========
'b'         blue
'g'         green
'r'         red
'c'         cyan
'm'         magenta
'y'         yellow
'k'         black
'w'         white
==========  ========
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
#사이즈만큰 랜덤한 데이터 생성 (정규분포: 평균0, 표준편차1)
y=np.random.standard_normal(size)
x=range(size)
# 차트 꾸미기
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정 # b = blue, lw =선의 굵기
plt.plot(x,y, 'o', ) #차트 내 'o' 표기
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
#사이즈만큰 랜덤한 데이터 생성 (정규분포: 평균0, 표준편차1)
y=np.random.standard_normal(size)
x=range(size)

# 차트 꾸미기
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정

```

## 차트 꾸미기 (실습: 어노테이션)

어노테이션


```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
x = np.linspace(1,10,10)
y = np.linspace(1,20,10)
y2= x**2

#차트 사이즈를 바꿔줄 수 있다.
plt.figure(figsize=(20,10))

# 차트 꾸미기
plt.plot(x,y, "b--", label="first") #차트 선색 및 크기설정
plt.plot(x,y2, "r--", label="second") #차트 선색 및 크기설정
plt.legend(loc=1)

# 화살표 시작지점
# '분리지점' 한글이 깨지면.. 
plt.annotate('분리지점',xy=(2,2), xytext=(2,20),arrowprops=dict(color="blue", width=2))
```

## 2차원 차트 Subplot


```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))

y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(121) #1행 2열에서 첫번째
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'o') # o라는 동그라미 점 같이 그려주려고 넣은것
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')

plt.subplot(122) #1행 2열에서 두번째
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go') # green  동그라미 점
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))

y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(211) #1행 2열에서 첫번째
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')

plt.subplot(212) #1행 2열에서 두번째
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

## 2차원 차트 그리기

Tableau 에서도 그리기 쉽지 않은데, 활용은 위에 Subplot이 더 많이 된다

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')

#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx() 
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

## 차트 종류 변경

라인, 파이, 스캐터를 많이 쓴다


```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 150개의 2열짜리 정규분포 생성
y=np.random.standard_normal((150,2))

plt.figure(figsize=(7,5))
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
plt.scatter(y[:,0],y[:,1],marker='o', edgecolors='r') # marker 0 동그라미로, red color로
#plt.scatter?
```

## 차원추가 (Scatter 차트)

하나더 늘려서 차원을 표시해주는것

스케터 많이 쓰인다. 기억

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 정규분포 50개행 컬럼2개
y=np.random.standard_normal((50,2))
# 0~10까지 랜덤숫자 50개
c= np.random.randint(0,10,len(y)) #0~10 가지 색깔 랜덤

plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o',c=c)
plt.colorbar()

plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
#plt.scatter?
```

## 객체 차트

서브로 차트를 하나 더 그릴 수 있다!

```python
#객체 차트
# 빈캔버스에 추가함
fig = plt.figure()
# 위치 및 너비/높이

# 1~5까지 11개의 점 삽입
x = np.linspace(1,5,11) 
y = x**2 # y = x의 제곱

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # x, y
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.3])

ax1.plot(x , y)
ax2.plot(y , x) #y, x 거꾸로.
ax1.set_title('plot1')
ax2.set_title('plot2')
fig
```
