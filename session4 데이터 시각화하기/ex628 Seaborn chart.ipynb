## Distplot 차트 구현

특정 컬럼 값의 분포도를 확인하기위한 차트

씨본 최근 뜨는데, 좀 더 이쁘고, 자체적으로 분석기능도 있다.


```python
#pip install seasborn
import seaborn as sns 
%matplotlib inline

tips = sns.load_dataset('tips') #기본 제공 tips data가 있다. iris 같은..
tips.head()


```

```python
#  tips data의  total_bill 에 대한 분포가 얼마나 되어있는지. 
# 20달러 정도에 몰려있다는걸 알 수 있음.

sns.distplot(tips['total_bill'])
```

## Jointplot 차트 구현

데이터셋에 대하여 특정 2가지 값에 대하여 그래프형태로 표현 

금액대비 실제 tip 분포 및
각 카테고리별 분포도를 함께 볼수 있음


실제 10달러 썼을때, 팁을 1달러주고..

10~20달러 분포,

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

## pairplot 차트 구현

전체 데이터의 모든관계를 한눈에 보기위한 차트


```python
sns.pairplot(tips)

# groupby = hue  
# sex에 male, female을 group by 한 데이터는 color로 표현해준다.
sns.pairplot(tips, hue = 'sex')
```

```python
import pandas as pd
from sqlalchemy import create_engine 
 
# DB 커넥션 열기
engine = create_engine('oracle+cx_oracle://kopo:kopo@192.168.110.112:1521/orcl') 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_sql_query('SELECT * FROM customerdata', engine) 

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']
customerData.head(2)
```

#### 2. np.where 숫자 바꾸기

```python
np.where 이것 같이. 데이터 조작하는건 알아두어야 한다
시각화 보단 데이터조작이 더 인정 받음. + 밑에 람다 함수 알아두어야
```

```python
import numpy as np
customerData["CUSTTYPE_NEW"] = np.where(customerData["CUSTTYPE"] == "Big-Screen-lover", 1,\
                                          np.where(customerData["CUSTTYPE"] == "Sleeping-dog", 2,\
                                          np.where(customerData["CUSTTYPE"] == "Early-bird", 3,\
                                          4)))
```

```python
customerData.head()
```

```python
람다 함수 알아두기
CUSTTYPE x마다 colors 에 접근하면서 해당 color key를 가져오는거다.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots()

colors = {'Big-Screen-lover':'red', 'Sleeping-dog':'blue', 'Early-bird':'green'}

ax.scatter(customerData.DEVICECOUNT, customerData.PRODUCTAGE, c=customerData.CUSTTYPE.apply(lambda x: colors[x]))
plt.show()
```

### 람다가 좀 어려우면, 아래 처럼 함수로 바꾸는 방법으로 알아두어도 된다

```python
# 정답은 아니고 이런식으로 한다는 것만.
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return df['DEVICECOUNT']
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5
```

```python
customerData["productage_new2"] =customerData.apply(refining, axis=1)
```
